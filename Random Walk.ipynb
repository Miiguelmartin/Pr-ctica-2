```python
# Una dimensi√≥n
import random
import matplotlib.pyplot as mat
random.seed()
distancia_0 = 1    
n = 100        
factor = 1.01   
prob = 0.5          
valoracion = [distancia_0] 
for t in range(1,n):
    if random.random() < prob:
        distancia_0 *= factor
    else:
        distancia_0 /= factor
    valoracion.append(distancia_0)
mat.plot(range(n),valoracion)
mat.show()
```

```python
# Dos Dimensiones
import random  
import math  
import matplotlib.pyplot as mat 
random.seed()  
n=30 
r=1  
x=y=0 
origenx=[x]  
origeny=[y]  
for i in range(n):  
  angulo=random.random()*math.pi*2 
  x+=r*math.cos(angulo)  
  y+=r*math.sin(angulo)  
  origenx.append(x)  
  origeny.append(y)  
mat.plot(origenx,origeny)  
mat.show()
```
